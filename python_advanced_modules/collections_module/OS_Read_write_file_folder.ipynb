```python
# os.walk(file_path) - Returns folder -> sub_folders -> files .. inside all
file_path = 'c:\\Users\\yogih\\My Python Workspace'
for folder, sub_folder, files in os.walk(file_path):
    print(f'Currently looking at {folder}')
    print('\n')
    print('The subfolders are :')
    for sub_folder in sub_folder:
        print(f"\tSubfolder: {sub_folder}")

    print('\n')
    print('The files are :')
    for f in files:
        print(f"\tFile : {f}")
    print('\n')
```
